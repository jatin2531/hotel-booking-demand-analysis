```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("hotel_booking_cleaned_data.csv")
df.head()
```

```python
# check info
df.info()
# check missing values
df.isnull().sum()
# remove missing values
df = df.dropna()
# remove duplicates
df.drop_duplicates(inplace=True)
# Convert string dates to datetime using correct format
df['Arrival Date'] = pd.to_datetime(df['Arrival Date'], format='%d-%b-%y')
df['Booking Date'] = pd.to_datetime(df['Booking Date'], format='%d-%b-%y')
#create new column for month
df['Month'] = df['Arrival Date'].dt.month
```

# - Univariate Analysis

```python
#Univariate Analysis of Booking Status Distribution
# Show both cleanly
fig, axes = plt.subplots(1, 2, figsize=(10,5))

# Countplot
ax = sns.countplot(x='Booking Status', data=df, ax=axes[0], color='steelblue')
axes[0].set_xticks([0,1])
axes[0].set_xticklabels(['Not Cancelled','Cancelled'])
for p in ax.patches:
    axes[0].annotate(f'{int(p.get_height())}',
                     (p.get_x() + p.get_width()/2, p.get_height()),
                     ha='center', va='bottom')
axes[0].set_title("Bar Diagram for Booking Status")
# Pie chart
counts = df['Booking Status'].value_counts()
axes[1].pie(counts,
            labels=['Not Cancelled','Cancelled'],
            autopct='%1.1f%%',
            colors=['steelblue','salmon'])
axes[1].set_title("Pie chart for Booking Status")
plt.legend()
plt.tight_layout()
plt.show()
```

### Booking Status Insight
Out of total bookings, 684(i.e 68.4%) bookings were not cancelled while 316 (i.e 31.6%)bookings were cancelled.

This indicates that the majority of customers complete their bookings successfully.
However, a significant portion (around 31.6%) of bookings are cancelled, which suggests
that cancellation is an important factor to consider in hotel booking analysis.

```python
#Univariate Analysis of Hotel Type Distribution
# Countplot to show number of bookings for each hotel type
ax = sns.countplot(x='Hotel', data=df)
# Add count labels on bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width()/2, p.get_height()),
                ha='center', va='bottom')

plt.title("Hotel Type Distribution")
plt.xlabel("Hotel Type")
plt.ylabel("Count")
plt.show()
```

### Hotel Type Insight
Out of total bookings, 627 bookings are for Resort Hotels while 373 bookings are for City Hotels.

This indicates that Resort Hotels are more preferred by customers compared to City Hotels.
The higher number of bookings for Resort Hotels suggests that customers are more interested
in leisure travel and vacation stays rather than city-based or business-related stays.

```python

#Univariate Analysis of Market Segment Distribution
# Countplot to show number of bookings for each market segment
ax = sns.countplot(x='Market Segment', data=df)
# Add count labels on bars
ax.bar_label(ax.containers[0])

plt.xticks(rotation=45)
plt.title("Market Segment Distribution")
plt.xlabel("Market Segment")
plt.ylabel("Count")
plt.show()
```

### Market Segment Insight
Out of total bookings, Online TA has the highest bookings (576), followed by Company TA/TO (209) and Direct bookings (182).
Other segments such as Groups (20) and Corporate (13) contribute very less to total bookings.

This indicates that most customers prefer booking through online travel agents, showing the importance of online platforms in hotel bookings, while corporate and group bookings are comparatively lower.

```python
#Distribution of Average Daily Rate (ADR)
# Histogram shows how ADR values are distributed across bookings
sns.histplot(df['Average Daily Rate'], bins=20, kde=True, color='purple')
plt.title("Distribution of Average Daily Rate (ADR)")
plt.xlabel("ADR")
plt.ylabel("Frequency")
plt.show()
```

### Average Daily Rate Insights
The ADR (Average Daily Rate) distribution is right-skewed, indicating that most bookings are concentrated in the mid-price range, mainly between ₹110 and ₹140. This represents the core pricing segment of the hotel business and shows that most customers prefer moderately priced stays.

A small number of bookings have very low or zero ADR, which may indicate complimentary stays or data anomalies, while a few high-value bookings above ₹180 represent premium segments. Overall, the pricing structure is stable, with most revenue generated from mid-range bookings and limited influence from extreme values.

```python
#Lead Time Distribution
# Histogram shows how far in advance customers book their rooms
df.columns = df.columns.str.strip()   #remove extra spaces
sns.histplot(df['Lead Time'], bins=20, kde=True, color='green')
plt.title("Lead Time Distribution")
plt.xlabel("Lead Time (Days)")
plt.ylabel("Number of Bookings")
plt.show()
```

### Lead Time Insights
The Lead Time distribution is strongly right-skewed, showing that most customers book their stays within a short period before arrival (0–50 days). This indicates a preference for last-minute or short-term booking behavior.

As lead time increases, the number of bookings decreases significantly, with very few long-advance bookings beyond 100 days and a few rare outliers above 300 days. Overall, the hotel mainly receives short-to-medium term bookings with limited long-term reservations.

```python
# Revenue calculation
df['revenue'] = df['Average Daily Rate'] * df['Total Nights Stay']

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14,10))

# ----------------------------
# Row 1 → Distribution (Pie Charts)
# ----------------------------

# Reserved Room Type Distribution
res_counts = df['Reserved Room Type'].value_counts()
axes[0,0].pie(res_counts, autopct='%1.1f%%', startangle=90,
              colors=['#66c2a5', '#fc8d62', '#8da0cb'])
axes[0,0].set_title("Reserved Room Type Distribution")

# Assigned Room Type Distribution
ass_counts = df['Assigned Room Type'].value_counts()
axes[0,1].pie(ass_counts, autopct='%1.1f%%', startangle=90,
              colors=['#66c2a5', '#fc8d62', '#8da0cb'])
axes[0,1].set_title("Assigned Room Type Distribution")

# ----------------------------
# Row 2 → Revenue (Bar Charts)
# ----------------------------

# Reserved Room Revenue
reserved_revenue = df.groupby('Reserved Room Type')['revenue'].sum()
axes[1,0].bar(reserved_revenue.index, reserved_revenue.values)
axes[1,0].set_title("Revenue by Reserved Room Type")

# Assigned Room Revenue
assigned_revenue = df.groupby('Assigned Room Type')['revenue'].sum()
axes[1,1].bar(assigned_revenue.index, assigned_revenue.values)
axes[1,1].set_title("Revenue by Assigned Room Type")

plt.tight_layout()
plt.show()
```

### Room Type Insights

The Reserved Room Type distribution shows that Standard rooms account for approximately 48.4% of total bookings, followed by Deluxe (~30%) and Premium (~21%). This indicates that customer demand is heavily concentrated in lower-priced room categories, reflecting price-sensitive booking behavior.

In contrast, the Assigned Room Type distribution is more balanced, with each room category contributing roughly 30–35% of total allocations. This suggests that a significant number of customers are being upgraded from Standard to higher room categories due to availability constraints.

While such upgrades can enhance customer satisfaction, they may also lead to potential revenue inefficiencies, as higher-value rooms are allocated without corresponding pricing adjustments. This highlights the need for improved room allocation and inventory management strategies to optimize revenue.

```python
# Box Plot of ADR vs Booking Status
df['Booking Status Label'] = df['Booking Status'].map({
    0: 'Not Cancelled',
    1: 'Cancelled'
})
plt.figure(figsize=(6,4))
order = ['Not Cancelled', 'Cancelled']
ax = sns.boxplot(
    x='Booking Status Label',
    y='Average Daily Rate',
    data=df,
    order=order,
    hue='Booking Status Label',   # fix for warning
    palette='Blues',
    legend=False
)
#for finding Q1,Q2,Q3
for i, label in enumerate(order):
    subset = df[df['Booking Status Label'] == label]['Average Daily Rate']
    q1 = subset.quantile(0.25)
    q2 = subset.quantile(0.50)
    q3 = subset.quantile(0.75)
    plt.text(i, q1, f'Q1:{q1:.0f}', ha='center', va='top', fontsize=8)
    plt.text(i, q2, f'Q2:{q2:.0f}', ha='center', va='bottom', fontsize=8)
    plt.text(i, q3, f'Q3:{q3:.0f}', ha='center', va='bottom', fontsize=8)
plt.title("ADR vs Booking Status")
plt.xlabel("Booking Status")
plt.ylabel("Average Daily Rate")
plt.show()
```

The boxplot shows that the Average Daily Rate (ADR) is almost similar for both cancelled and not cancelled bookings. The median value and the overall price range are nearly the same in both groups, which indicates that pricing does not have a strong impact on whether a booking is cancelled or not.

There are a few outliers in both groups, including very low ADR values (around 0) and some very high-priced bookings. These represent unusual or special cases, but they do not significantly affect the overall pattern. Overall, cancellations are not mainly driven by differences in price.

```python
# Boxplot of lead time vs booking status

df['Booking Status Label'] = df['Booking Status'].map({
    0: 'Not Cancelled',
    1: 'Cancelled'
})
plt.figure(figsize=(6,4))
order = ['Not Cancelled', 'Cancelled']
sns.boxplot(
    x='Booking Status Label',
    y='Lead Time',
    data=df,
    order=order,
    hue='Booking Status Label',   # fix
    palette='Oranges',
    legend=False
)
#for finding Q1,Q2,Q3
for i, label in enumerate(order):
    subset = df[df['Booking Status Label'] == label]['Lead Time']

    q1 = subset.quantile(0.25)
    q2 = subset.quantile(0.50)
    q3 = subset.quantile(0.75)

    plt.text(i, q1, f'Q1:{q1:.0f}', ha='center', va='top', fontsize=8)
    plt.text(i, q2, f'Q2:{q2:.0f}', ha='center', va='bottom', fontsize=8)
    plt.text(i, q3, f'Q3:{q3:.0f}', ha='center', va='bottom', fontsize=8)
plt.title("Lead Time vs Booking Status")
plt.xlabel("Booking Status")
plt.ylabel("Lead Time")
plt.show()
```

The boxplot shows that lead time is almost the same for both cancelled and not cancelled bookings, so booking early or late does not strongly affect cancellations. Both groups have similar median and spread, with only a few extreme outliers where very early bookings sometimes get cancelled.
Overall,lead time is not a strong factor in cancellations in this dataset.

```python
#Market segment vs cancellation rate analysis
df['Booking Status Label'] = df['Booking Status'].map({
    0: 'Not Cancelled',
    1: 'Cancelled'
})
cancel_rate = df.groupby('Market Segment')['Booking Status'].mean() * 100
cancel_rate = cancel_rate.sort_values(ascending=False)
plt.figure(figsize=(8,5))
ax = cancel_rate.plot(kind='bar', color='salmon')
for i, v in enumerate(cancel_rate):
    plt.text(i, v + 1, f'{v:.1f}%', ha='center', fontsize=9)

# Titles and labels
plt.title("Cancellation Rate by Market Segment (%)")
plt.xlabel("Market Segment")
plt.ylabel("Cancellation Rate (%)")
plt.xticks(rotation=30)
plt.ylim(0, max(cancel_rate) + 5)

plt.tight_layout()
plt.show()
```

The chart shows that the booking channel has a clear impact on cancellations. Online TA records the highest cancellation rate, indicating a higher risk associated with online bookings. In comparison, Direct bookings have a much lower cancellation rate, making them more stable. Corporate and Offline TA/TO segments also show relatively lower cancellations.

Overall, the likelihood of cancellation varies across booking channels, with online platforms being the most prone to cancellations, while direct and corporate bookings are more dependable.

# -Customer Segmentation

In this section, we group customers based on booking behavior such as lead time and stay duration.

```python
#Customer Segmentation – Lead Time Distribution
# Create Lead Time Segments (3 groups)
df['Lead Time Segment'] = pd.cut(
    df['Lead Time'],
    bins=[0, 7, 60, 365],
    labels=['Last Minute', 'Medium Planner', 'Early Planner']
)
# Plot graph
plt.figure(figsize=(6,4))
sns.countplot(
    x='Lead Time Segment',
    data=df,
    hue='Lead Time Segment',
    palette='Set2',
    legend=False,
    dodge=False   # fix bar size
)
plt.title("Customer Segmentation - Lead Time")
plt.xlabel("Segment")
plt.ylabel("Count")
plt.show()
```

Most customers are medium planners, booking their stays 1–3 months in advance, forming the main booking segment of the hotel. A noticeable portion of customers are last-minute bookers, indicating strong short-term demand, while early planners represent a smaller segment.
Overall, the booking pattern is stable and predictable, with opportunities to apply early-bird pricing for advance bookings and dynamic pricing for last-minute demand.

```python
#Customer Segmentation- Stay duration
def stay_segment(x):
    if x <= 3:
        return "Short Stay"
    elif x <= 7:
        return "Medium Stay"
    else:
        return "Long Stay"
df['Stay Segment'] = df['Total Nights Stay'].apply(stay_segment)
plt.figure(figsize=(6,4))
sns.countplot(
    x='Stay Segment',
    data=df,
    hue='Stay Segment',   # fixes warning
    palette='cool',
    legend=False
)
plt.title("Customer Segmentation - Stay Duration")
plt.xlabel("Stay Type")
plt.ylabel("Count")
plt.show()
```

The chart shows that most customers are short-stay guests, staying only 1–3 nights, making this the main customer group for the hotel. A smaller portion are medium-stay guests, while very few customers stay for long durations.

Overall, the hotel mainly serves quick trips and short visits, with high guest turnover and limited long stays. This suggests focusing on short-stay demand while also encouraging medium-stay packages to increase revenue.

# - Monthly Trend and Seasonality Analysis

```python
df.columns = df.columns.str.strip()

df['Month'] = pd.to_datetime(df['Arrival Date']).dt.strftime('%b')
month_order = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
#GRAPH 1 : Monthly Booking VS Cancellation Trend
monthly_data = df.groupby('Month')['Booking Status'].value_counts().unstack().fillna(0)
monthly_data = monthly_data.reindex(month_order)

#plot
fig, axes = plt.subplots(1, 2, figsize=(18,6))


monthly_data.plot(kind='bar', ax=axes[0], color=['steelblue','salmon'])
axes[0].set_title("Monthly Booking vs Cancellation Trend")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Number of Bookings")
axes[0].legend(['Not Cancelled','Cancelled'])
axes[0].tick_params(axis='x', rotation=45)

# GRAPH 2 : Monthly Trend (Peak/Low Season)

month_counts = df['Month'].value_counts().reindex(month_order).fillna(0)
peak_months = month_counts.nlargest(3).index
low_months = month_counts.nsmallest(3).index

axes[1].plot(month_counts.index, month_counts.values,
             marker='o', color='black', label="Bookings")
#Highlighting with colors
for m in peak_months:
    if m in month_counts.index:
        axes[1].scatter(m, month_counts[m],
                        color='red', s=120,
                        label='Peak Season' if m == peak_months[0] else "")

for m in low_months:
    if m in month_counts.index:
        axes[1].scatter(m, month_counts[m],
                        color='blue', s=120,
                        label='Low Season' if m == low_months[0] else "")

axes[1].set_title("Peak vs Low Season Analysis ")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Number of Bookings")
axes[1].tick_params(axis='x', rotation=45)
axes[1].legend()
axes[1].grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

## Monthly Booking & Seasonality Analysis

 ### - Monthly Booking vs Cancellation Trend

June shows the highest booking volume, indicating peak demand and strong travel activity. November records the highest cancellations, making it a critical month despite high booking activity. March and May show relatively lower booking activity, indicating weaker demand periods.

Overall, high-booking months like June and November also experience higher cancellations, showing that cancellation behavior is more significant during peak demand periods and requires better booking management strategies.

 ### - Peak vs Low Season Trend

The seasonality analysis highlights clear demand fluctuations across the year.

Peak Season Months (Top 3): June, January, November .
These months show the highest booking activity, marked in red, indicating strong seasonal demand and maximum hotel occupancy.

Low Season Months (Bottom 3): March, April, May .
These months show comparatively lower booking counts, marked in blue, reflecting reduced demand and weaker travel activity.

This shows clear seasonality in hotel demand, with mid and year-end months performing better than early-year months.


```python
#Monthly ADR vs cancellation rate
df['Month'] = pd.to_datetime(df['Arrival Date']).dt.strftime('%b')
month_order = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
monthly_adr = df.groupby('Month')['Average Daily Rate'].mean().reindex(month_order)
monthly_cancel = df.groupby('Month')['Booking Status'].mean().reindex(month_order)
#ADR line
fig, ax1 = plt.subplots(figsize=(12,6))
ax1.plot(monthly_adr.index, monthly_adr.values, marker='o', linewidth=2,  label='ADR')
ax1.set_xlabel("Month",fontsize=11)
ax1.set_ylabel("Average Daily Rate",fontsize=11)
ax1.set_title("Monthly ADR vs Cancellation Rate Trend", fontsize=14, fontweight='bold')
ax1.grid(True, linestyle='--', alpha=0.5)
top3_adr = monthly_adr.sort_values(ascending=False).head(3)
ax1.scatter(top3_adr.index, top3_adr.values, color='red', s=120, label='Top ADR Months')
#Cancellation line
ax2 = ax1.twinx()
ax2.plot(monthly_cancel.index, monthly_cancel.values, marker='o', linestyle='--', linewidth=2,  label='Cancellation Rate')
ax2.set_ylabel("Cancellation Rate", fontsize=11)
top3_cancel = monthly_cancel.sort_values(ascending=False).head(3)
ax2.scatter(top3_cancel.index, top3_cancel.values, color='red', s=120, label='High Cancellation Months')
#Annotate values
for i, v in enumerate(monthly_adr):
    if not pd.isna(v):
        ax1.text(i, v, f'{v:.0f}', ha='center', va='bottom', fontsize=8)

for i, v in enumerate(monthly_cancel):
    if not pd.isna(v):
        ax2.text(i, v, f'{v:.2f}', ha='center', va='top', fontsize=8)
#combining
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

plt.tight_layout()
plt.show()



```

### Monthly ADR vs Cancellation Rate Insights

The Monthly ADR vs Cancellation Rate Trend shows clear seasonal variation in both pricing and cancellations. February and October record the highest ADR (~140), indicating peak demand periods, while March shows the lowest ADR (~129), reflecting lower demand. Cancellation rates also vary across months, with certain months showing relatively higher cancellation levels compared to others.

However, there is no consistent relationship between ADR and cancellation rates, as months with higher pricing do not always correspond to higher cancellations. This indicates that customers do not cancel bookings primarily due to price changes.

Overall, while ADR follows a clear seasonal trend driven by demand, cancellations are influenced more by other factors such as booking channel, customer behavior, and planning patterns rather than pricing alone.

# Correlation Analysis

### Key Insights

* Strong relationship between **Total Guests and Children (~0.89)**, indicating a high proportion of family-based bookings.

### Weak Relationships

* **Average Daily Rate (ADR)** shows very weak correlation with Booking Status → pricing is not a major driver of cancellations.

* **Room Change** has negligible correlation → reassignment or upgrades do not significantly influence cancellations.



### Conclusion

* The strongest correlations are mainly between **booking structure variables** (guests, nights, children), not cancellations.
* **Cancellation behavior is weakly correlated with most features**, indicating it is influenced by multiple factors rather than a single dominant variable.
* Overall, **customer behavior and booking patterns** play a more important role than pricing or timing in determining cancellations.
